```python
import numpy as np

# essai moyenne exponentielle mobile

n = 100
data = np.random.normal(loc=7.5, scale=2, size=n).astype(int)
improve = np.round(np.linspace(0, 2, 100),0)
data = data + improve

import matplotlib.pyplot as plt

plt.hist(data, bins=10, edgecolor='black')
plt.title('Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(data, ".")
ax.set_title('Evolution of Data')
ax.set_ylim(0, 10)
ax.set_xlabel('Index')
ax.set_ylabel('Value')

period = 19
alpha = 2 / (period + 1)

print(alpha)

def exponential_moving_average(data, alpha=0.1):
    ema = [data[0]]
    for i in range(1, len(data)):
        ema.append(alpha * data[i] + (1 - alpha) * ema[i - 1])
    return ema

ema = exponential_moving_average(data, alpha)
ax.plot(ema, label='Exponential Moving Average', color='red')
```
